```python
import cv2 
import matplotlib.pyplot as plt

#reading image
img1 = cv2.imread('../data/oxbuild_images-v1/all_souls_000006.jpg')  
gray1 = cv2.cvtColor(img1, cv2.COLOR_BGR2GRAY)

#keypoints
sift = cv2.xfeatures2d.SIFT_create()
keypoints_1, descriptors_1 = sift.detectAndCompute(img1,None)

img_1 = cv2.drawKeypoints(gray1,keypoints_1,img1)
plt.imshow(img_1)
```

```python
img1 = cv2.imread('../data/oxbuild_images-v1/all_souls_000006.jpg')  
img2 = cv2.imread('../data/oxbuild_images-v1/all_souls_000160.jpg') 

img1 = cv2.cvtColor(img1, cv2.COLOR_BGR2GRAY)
img2 = cv2.cvtColor(img2, cv2.COLOR_BGR2GRAY)

figure, ax = plt.subplots(1, 2, figsize=(16, 8))

ax[0].imshow(img1, cmap='gray')
ax[1].imshow(img2, cmap='gray')
```

```python
#sift
sift = cv2.xfeatures2d.SIFT_create()

keypoints_1, descriptors_1 = sift.detectAndCompute(img1,None)
keypoints_2, descriptors_2 = sift.detectAndCompute(img2,None)

len(keypoints_1), len(keypoints_2)
```

```python
bf = cv2.BFMatcher(cv2.NORM_L1, crossCheck=True)

matches = bf.match(descriptors_1,descriptors_2)
matches = sorted(matches, key = lambda x:x.distance)

img3 = cv2.drawMatches(img1, keypoints_1, img2, keypoints_2, matches[:50], img2, flags=2)
plt.imshow(img3),plt.show()
```

```python
import os 
from tqdm import tqdm
from numpy import save, load
os.listdir('../data/oxbuild_images-v1')[:5]
```

```python
des = {}

for img2_path in tqdm(os.listdir('../data/oxbuild_images-v1')):
    if not img2_path.endswith("jpg"):
        continue
    img2 = cv2.imread('../data/oxbuild_images-v1/' + img2_path)
    img2 = cv2.cvtColor(img2, cv2.COLOR_BGR2GRAY)
    keypoints_2, descriptors_2 = sift.detectAndCompute(img2,None)
    des[img2_path] = descriptors_2
```

```python
def find_best(img_path):
    descriptors_1 = des[img_path]
    # bf = cv2.BFMatcher(cv2.NORM_L1, crossCheck=True)
    
    # FLANN parameters
    FLANN_INDEX_KDTREE = 1
    KKK = 2
    index_params = dict(algorithm = FLANN_INDEX_KDTREE, trees = 5)
    search_params = dict(checks=50)   # or pass empty dictionary
    flann = cv2.FlannBasedMatcher(index_params,search_params)
    
    md = {}
    for img2_path in tqdm(os.listdir('../data/oxbuild_images-v1')):
        if not img2_path.endswith("jpg"):
            continue
        # if not img2_path.startswith("all_souls"):
        #     continue
        descriptors_2 = des[img2_path]
        if (not descriptors_2 is None) and (not len(descriptors_2)<KKK):
            # matches = bf.match(descriptors_1,descriptors_2)
            
            knn_matches = flann.knnMatch(descriptors_1,descriptors_2,k=KKK)
            ratio_thresh = 0.7
            good_matches = []
            for m,n in knn_matches:
                if m.distance < ratio_thresh * n.distance:
                    good_matches.append(m)
                    
            md[img2_path] = len(good_matches)
        else:
            md[img2_path] = 0
    return md
```

```python
temp = find_best('all_souls_000006.jpg')
```
